```
def print_most_similar_jokes(ratingsnp, jokes, query_joke_id, k, metric=recommender.pearsSim):
    print 'query joke text: \n', jokes[query_joke_id][0]
    print 'Top', k, 'jokes we recommend are :'
    ratingT = ratingsnp.T
    sim = []
    for i in range(ratingT.shape[0]):
        dist=metric(ratingT[query_joke_id], ratingT[i])
        dist2=[dist, i]
        sim.append(dist2)
    sim.sort(reverse=True)
    sim = sim[1:10+1]
    for i in range(len(sim)):
        print 'Joke #',sim[i][1],":\n",jokes[sim[i][1]][0]
    return sim
```
